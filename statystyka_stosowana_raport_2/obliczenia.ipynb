```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from scipy import stats
from numba import jit
```

# Zadanie 1

```python
mu_0 = 1.5
sigma = 0.2
```

```python
data = np.loadtxt("zad1.txt", dtype = float)
n = len(data)
```

```python
Z = (np.mean(data) - mu_0)/(sigma/np.sqrt(n))
Z
```

### $\mu \neq 1.5$

```python
p_value = 2*(1 - stats.norm.cdf(np.abs(Z), loc=0, scale=1))
print(f"p-wartość: {p_value}")
```

```python
alpha = 0.05
z_half_alpha = stats.norm.ppf(1 - alpha/2, loc = 0, scale = 1)

# punkty krytyczne:
print(f'Punkty krytyczne dla alpha = {alpha}: {-z_half_alpha} oraz {z_half_alpha}')

# obszary krytyczne mają postac: (-infinity, -1.959963984540054] and [1.959963984540054, infinity)
```

```python
x = np.linspace(-4, 4, n)
f_theo = stats.norm.pdf(x, 0, 1)

plt.plot(x, f_theo, c = 'black')

x_1 = np.linspace(-4, -z_half_alpha, 200)
x_2 = np.linspace(z_half_alpha, 4, 200)

plt.fill_between(x=x_1, y1=stats.norm.pdf(x_1, 0, 1), color='#1f77b4', alpha=.8)
plt.fill_between(x=x_2, y1=stats.norm.pdf(x_2, 0, 1), color='#1f77b4', alpha=.8)

plt.axvline(-z_half_alpha, c = '#1f77b4', linestyle = '-', alpha=.9)
plt.axvline(z_half_alpha, c = '#1f77b4', linestyle = '-', alpha=.9)

plt.text(-2.8, 0.2, "-1.960", color='#1f77b4')
plt.text(2, 0.2, "1.960",color='#1f77b4')

plt.title(f"Obszary krytyczne dla $\\alpha =${alpha} (przypadek $H_{1}:\\mu \\neq 1.5$)")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend(frameon=False)
```

```python
sigma_0 = 1.5 # podniesiona do kwadratu
mu = 0.2
```

```python
data = np.loadtxt("zad2.txt", dtype = float)
n = len(data)
```

```python
Chi_2 = sum((data - np.mean(data))**2)/sigma_0
Chi_2
```

### $\sigma^{2} \neq 1.5$

```python
p_value = 2*(1 - stats.chi2.cdf(np.abs(Chi_2), n-1, loc=0, scale=1))
print(f"p-wartość: {p_value}")
```

```python
alpha = 0.05
x_half_alpha_1 = stats.chi2.ppf(1 - (alpha/2), n - 1)
x_half_alpha_2 = stats.chi2.ppf((alpha/2), n - 1)

# punkty krytyczne:
print(f'Punkty krytyczne dla alpha = {alpha}: {x_half_alpha_1} oraz {x_half_alpha_2}')

# obszary krytyczne mają postac: (-infinity, 913.3009983021134] oraz [1088.4870677259353, inifinity)
```

```python
x = np.linspace(800, 1200, n)

f_theo = stats.chi2.pdf(x, n - 1, 0, 1)
plt.plot(x, f_theo, c = 'black')

x_1 = np.linspace(800, x_half_alpha_2, 200)
x_2 = np.linspace(x_half_alpha_1, 1200, 200)

plt.fill_between(x=x_1, y1=stats.chi2.pdf(x_1, n - 1, 0, 1), color='#1f77b4', alpha=.8)
plt.fill_between(x=x_2, y1=stats.chi2.pdf(x_2, n - 1, 0, 1), color='#1f77b4', alpha=.8)

plt.axvline(x_half_alpha_2, c = '#1f77b4', linestyle = '-', alpha=.9)
plt.axvline(x_half_alpha_1, c = '#1f77b4', linestyle = '-', alpha=.9)

plt.text(1092, 0.004, "1088", color='#1f77b4')
plt.text(887, 0.004, "913",color='#1f77b4')

plt.title(f"Obszary krytyczne dla $\\alpha =${alpha} (przypadek $H_{1}:\\sigma^{2} \\neq 1.5$)")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend(frameon=False)
```
